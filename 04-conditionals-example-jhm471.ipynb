```python
# This script simulates playing a simple dice game, "Knock Out". The object of the game is to be the first to reach a chosen score
# by rolling a pair of dice. A knock out number is randomly assigned at the beginning of play, which, if rolled, will not contribute
# to a player's overall score. 

# Import random to simulate die rolls. Import time to seed random number generation differently for each game. Import sys to exit 
# program is user chooses.
import random
import time
import sys

# Prompt user to select score to be played to.
while True:
    try:
        score = int(input("What score would you like to play to?(enter number)")) # Score must be entered using number keys.
        break
    except ValueError: # If number keys are not used. Error message will be generated and user is reprompted for score.
        print("Sorry. Score must be entered using number keys.")

# Set scores of user and computer to zero.
user_score = 0
comp_score = 0

# Set random number seed using time.
random.seed(time.time())

# Assign knock out numbers to user and computer using random number generation.
user_ko = random.randint(1,6)
comp_ko = random.randint(1,6)
print("\n")
print("Your knock out number is ", str(user_ko))
print("Computer's knock out number is ", str(comp_ko))

while (user_score < score) and (comp_score < score): # Game set to end when score is met or surpassed by user or computer.
    
    # Prompt user to stop or continue game.
    roll = str(input("Do you want to roll dice?(y/n)"))
    
    # Reprompt user to roll if 'y' or 'n' is not entered.
    if (roll != 'y') and (roll != 'n'):
        print("\n")
        print("Sorry. You must enter 'y' or 'n'.")
        
    # If user types "n", script will close.
    elif roll == "n":
        print("\n")
        print("OK, quitting game.")
        sys.exit()
    
    # If user types "y", game will continue to be played.
    elif roll == "y":
        # Simulate dice rolls for user and computer using random number generation.
        user_die1 = random.randint(1,6)
        user_die2 = random.randint(1,6)
        comp_die1 = random.randint(1,6)
        comp_die2 = random.randint(1,6)
        
        # Notify user of user dice rolls and if knock out number was rolled.
        print("\n")
        print("You rolled ", str(user_die1), " and ", str(user_die2))
        if user_die1 == user_ko:
            print("Sorry, you rolled a knock out number on die 1. ", str(user_ko), " will not be added to your score.")
        if user_die2 == user_ko:
            print("sorry, you rolled a knock out number on die 2. ", str(user_ko), " will not be added to your score.")
        
        # Notify user of computer dice rolls and if knock out number was rolled.
        print("Computer rolled ", str(comp_die1), " and ", str(comp_die2))
        if comp_die1 == comp_ko:
            print("Computer rolled a knock out number on die 1. ", str(comp_ko), " will not be added to computer score.")
        if comp_die2 == comp_ko:
            print("Computer rolled a knock out number on die 2. ", str(comp_ko), " will not be added to computer score.")
            
        # Add dice rolls to total score and notify user of user score and computer score.
        if user_die1 != user_ko:
            user_score = user_score + user_die1
        if user_die2 != user_ko:
            user_score = user_score + user_die2
        if comp_die1 != comp_ko:
            comp_score = comp_score + comp_die1
        if comp_die2 != comp_ko:
            comp_score = comp_score + comp_die2
        print("\n")
        print("Your current score is ", str(user_score))
        print("Computer's current score is ", str(comp_score))
        
# Announce winner of game to user.
if user_score >= score:
    print("Congratulatioons! You won.")
elif comp_score >= score:
    print("Sorry. Computer won.")
```
